```python
#the examples from the textfile provided

def naivelog(x, n=15):
    #the parameter n is how many steps we take. to make question 2 easier i made it an argument
    #the bigger n, the better the approximation ln(1+x) = x is.
    #but the longer it takes and the more operations that introduce error.
  y = x
  for i in range(n): #repeat the tabbed code n times
    y = math.sqrt(y) #make y close to 1 to allow use of approximation. 
      #by taking square root it works for x in (0,1) and in (1, inf)
  l = y-1  #ln(1+x)~x for small x. 
    #this step causes cancellation of digits as we subtract two nearby numbers.
  for i in range(n):
    l = l*2 # ln(sqrt(sqrt(..(x)))) = 1/2*1/2*...1/2(ln(y)). 
      #so this step undoes the square rooting.
  return l

    
def mylog(x, n = 15): #this is a modification of naive log to remove cancellation of digits
  
  z = x-1 # we first do the subtraction.
  for i in range(n):
    z = z/(1+math.sqrt(1+z)) #here we make change z by making x closer to one   z = x-1 -> znew = sqrt(x) -1,
      #rearranged to avoid cancellation of digits.
  l = z #ln(1+a) = a => ln(z+1) = z = x-1 = ln(x)
  for i in range(n):
    l = l*2 # and finally use log laws as before to undo shrinking x.
  return l

```

```python
print(mylog(x),naivelog(x),math.log(x))
```

the following cells make and test functions to calculate exponentials, sines, and cosines.

```python
# calculating exponentials by argument halving
def my_exp(x):
    n = 15 
    y = x
    for i in range(n): #halve argument to make close to zero
        y = y/2        
    l = 1+y             # Use approximation
    for i in range(n): 
        l = l*l         # e^2x  = (e^x)^2 so repeating this undoes the halving.
    return l
```

This is implementing the example shown in class. We are using $e^{2x} = (e^x)^2$ for the argument halving portion, the approximation $e^x \approx 1+x$ for small x to get an approximation, and then the other side of the identity to obtain $e^x$

```python
for i in range(-10,11):
    print(i,my_exp(i), math.exp(i))
for i in (-100, -50, 50, 100, 500):
    print(i,my_exp(i), math.exp(i))
    
```

As you can see my method is pretty effective across scales, correct to 7 digits close to zero, and 4-5 digits for extreme values, even with the relatively small n used.

```python
#now sine. this is a straightforward implementation of work done in class. 
  
def my_sine(x):
    n = 15
    y = x
    for i in range(n):#we halve number n times to make it small, 
        y = y/2
    l = y             #use the small angle approximation to estimate its sine,
    for i in range(n):
        l = 2*l*math.sqrt(1-(l*l)) #and undo the halving with the double angle identity
        
    return l   #to return the sine of the original angle
```

This replicates class work. 
The key facts used are $\sin(x) \approx x$ for small x and the identity $\sin(2\theta) =  2\sin(\theta)\sqrt{1-\sin^2(\theta)}$ is used to enable the argument halving step

We compare results of my algorithm with the built in values.

```python
for i in range(-15, 19, 2):
    x = i/10
    print(x, my_sine(x), math.sin(x))
```

And we see that it is quite effective, being accurate in 9-11 digits using n = 15. The answer is accurate for the most digits close to 0, however is still very servicable outside of that range

We now do the same thing for cosine,using the identity $$\cos 2\theta = 2\cos^2 \theta -1$$ to enable argument halving and the cosine small angle approximation $$\cos \theta \approx 1 - \frac{1}{2}\theta^2$$ to approximate cosine once the argument has been made small.


```python
def my_cosine(x):
    n = 15
    y = x
    for i in range(n):
        y = y/2
    l = 1 - y*y*0.5 #small angle approx
    for i in range(n):
        l = 2*l*l - 1 #double angle identity
    return l
```

```python
for i in range(0, 40, 2):
    x = (i-8)/10
    print(x, my_cosine(x), math.cos(x))
```

And we see again that it is quite effective, although not quite as accurate as the algorithm for $\sin\theta$ I suspect the reason is that the subtraction in the cosine double angle identity causes some loss of digits, with the same thing potentially happening in the small angle approximation.

We implement a similar method as before using to get the solution. We will use the result shown in question 6b of  the first assignment to shrink x, then use a small angle approximation of arctan(x), and finally double to get the final result.
The identity used for the "argument halving" is
$$\frac{1}{2}\arctan(x) = \arctan \Biggl( \frac{x}{1+\sqrt{x^{2}+1}} \Biggr)$$
and for the small angle we use $$\tan(\theta) \approx \theta \implies \theta \approx \arctan(\theta)$$

```python
def my_arctan(x):
    n = 30 # make n big enough to have 10 digits correct for accuracy in part 2.
    y = x
    for i in range(n):
        y = y/(1+(math.sqrt(y*y + 1))) #shrink with identity
    l = y
    for i in range(n):
        l = 2*l
    return l
    
    
    
```

**General outline of Method**
I break down the problem into the following steps.
1. Create a lookup table of Arctangents of powers of 10. Call it ArctanValues
2. Create an algorithm called Cordic that performs pseudo-division and multiplication to rotate a vector up to an angle of $\theta$. We require at this step that $0<\theta<\frac{\pi}{2}$
3. Create an algorithm that calculates tangents using the vector generated by Cordic, called CordicTan
4. Create an algorithm that transforms arbitrary angles into an angle in the range $[-\pi, \pi]$ by adding and subtracting multiples of $2\pi$, called inPlusMinusPi
5. Create an algorithm that transformes the angle created by inPlusMinusPi into and angle $0<\theta<\frac{\pi}{2}$, and notes whether the final answer will need to be multiplied by -1. This is called TanReduc
6. Finally the algorithms are strung together to calculte the tangent of an arbitrary angle. This function is called FullCordicTan


```python
ArctanValues = {}
for i in range(0,5): # I will use 5 k-values the same as in the extra reading provided.
    
    value = 10**(-i)
    ArctanValues[value] = my_arctan(value)
print(ArctanValues)
```

1. The function Cordic I create in the following cell first uses pseudo division to break up the angle as  $$\theta = k_0\theta_0 + k_1\theta_1 + k_2\theta_2 + k_3\theta_3 + k_4\theta_4 + \theta_r $$ where $$\tan\theta_i = 10^{-i}$$
After this is completed we use the approximation $\sin\theta_r \approx \theta_r$ to create the y value of the starting vector and the approximation $\sin\theta_r \approx 1$ to create the x value.
We then rotate the vector (allowing it to stretch) as $$X_2 = X_1 + Y_1 \tan\theta$$ $$Y_2 = Y_1 - X_1 \tan\theta$$ Because we divided up the angle into convenient value of theta this reduces to $$X_2 = X_1 + Y_1 * 10^{-i}$$ $$Y_2 = Y_1 - X_1 *10^{-i}$$ After rotating by the full angle step by step we return the rotated vector

```python
def Cordic(theta):
    ks = [] #create a list to place the k-values in
    #this first loop takes theta and breaks it into a sum of angles with tangents that are powers of 10
    for i in range(0,5):
        key = 10**(-i) #generates the keys used in the dictionary that access the appropriate arctans
        k = 0 #We start counting how many arctans can fit in at 0.
        while theta- ArctanValues[key] > 0: # see if we can subtract another arctan(10^-i)
            theta = theta- ArctanValues[key]  #if yes we do
            k = k+1           #increase the count
            #and try again
        ks.append(k) #add our k value to the list
    #Now we set our starting vector
    y = theta
    x = 1
    #And rotate it using the algorithm shown in class.
    for i in range(0,5):
        for j in range(ks[i]):
            xnew = x - y*(10**(-i))
            ynew = y + x*(10**(-i))
            x = xnew  #do it like this rather than x = x - y*(10**(-i)) directly to avoid overwriting
            y = ynew
    return x, y

    
    
    
```

2. For an vector (x,y) that has an angle of $\theta$ with the origin vector (1,0), $$\tan\theta = \frac{y}{x}$$ this is wrapped around the Cordic function to obtain the tangent of the original angle.

```python
def CordicTan(theta):
    
    x, y = Cordic(theta)
    return y/x
    
```

3. We first deal with the case x = 0, then use the base 10 logarithm to determine the size of the angle. Based on this multiples of $2\pi$ are added and subtracted (so if our startin angle is 800, we start by subtracting 100s of $2/pi$s. Once the number is as close to zero as can be obtained by subtracting that multiple of $2\pi$ we move on to a lower multiple until we are in the range $[-\pi,\pi]$

```python
def inPlusMinusPi(x):  
    #get it within (-pi,pi) by subtracting 2pi * decreasing powers of ten
    if x == 0: 
        return x
    n = math.floor(math.log10(abs(x)))+1 
    for i in range(n)[::-1]:
        while abs(x)> math.pi*(10**i):       
            if x > math.pi*(10**i):
                x = x-2*math.pi*(10**i)
            elif x<-math.pi*(10**i):
                x = x + 2*math.pi*(10**i)
    return(x)
    
        
```

4. now we add and subtract $\pi$ and reflect about the origin to get $\theta$ in $[0,\pi/2]$ we note wether or not we will need to multiply the tangent of the transformed value by -1 to obtain the tangent of the original value. How this was done was worked out by looking at a graph of $\tan\theta$

```python
def TanReduc(theta):
     x = inPlusMinusPi(theta)
     if x<= -math.pi/2:
        x = x+ math.pi
        CorrectSign = True
     elif -math.pi/2<x<=0:
        x = -x
        CorrectSign = False
     elif 0<x<= math.pi/2:
        x = x
        CorrectSign = True
     else: #math.pi/2 <x<math.pi)
        x = -(x-math.pi)
        CorrectSign = False
     return (x, CorrectSign)
```

```python
print(TanReduc(-2000000*math.pi + 17))
```

5. finally we wrap it all together, passing the transformed angle to CordicTan, and applying the neccessary  correction.

```python
 def FullCordicTan(theta):
     TransformedTheta, CorrectSign = TanReduc(theta)
     Value = CordicTan(TransformedTheta)
     if CorrectSign:
         return Value
     else:
         return -Value
    
```

```python
for i in range(-5, 5):
    x = i/10
    print(x, FullCordicTan(x), math.tan(x))
for i in range(-5, 5):
    x = math.pi/2 + i/100
    print(x, FullCordicTan(x), math.tan(x))
print(math.pi/2, FullCordicTan(math.pi/2), math.tan(math.pi/2))
for x in (-1000, 100, 1000):
    print(x, FullCordicTan(x), math.tan(x))
```

So we see that the tangent function implemented works very well for values of $theta$ for which $\tan\theta$ is well behaved, having around 14 digits correct. $\tan\theta$ is badly behaved around $\pi/2$ however and there it disagrees entirely with the default.

```python
def CordicSin(theta):
    x,y = Cordic(theta)
    sin = y/math.sqrt(x**2 +y**2)
    return sin
def CordicCos(theta):
    x,y = Cordic(theta)
    cos = x/math.sqrt(x**2 + y**2)
    return cos
```

We follow the proceedure outlined in the Calculator reading.
arctangent takes a slope y/x = A, and returns the angle. Given A then we create a vector Y = A, X = 1, and rotate it by convenient increments of theta(allowing the length to change) using $$X'_2 = X_1 + Y_1 \tan\theta$$ $$Y'_2 = Y_1 - X_1 \tan\theta$$
We use the same trick as used in question 1, and rotate  the vector downwards towards (0,1) using increments of $\theta$ such that $\tan\theta = 1, 0.1,...$. this is simple pseudo division. We then use the table generated in question 1 to turn this pseudo quotient into a real angle.
The final consideration is that we will only calculate for positive values of A. This is fine since $\arctan(-A) = -\arctan(A)$ so we just need to take the absolute value of A and remember the original sign to fix at the end.

```python
def CordicArctan(A):
    CorrectSign = True
    if A< 0:
        A = abs(A)
        CorrectSign = False #note if we had to remove a negative sign
    Y = A
    X = 1
    ks = []
    for i in range (0, 5):                #divide it up by counting the angle rotated untill its right.
        k = 0
        while Y- X*(10**(-i))>0:
            Xnew = X + Y*(10**(-i))
            Ynew =  Y - X*(10**(-i))
            X = Xnew
            Y = Ynew
            k = k + 1
        ks.append(k)

    #turn the sum of little angles into a single angle
    #starting with the remainder.
    theta = Y/X        #this is using the small angle approximation where sin(y/x) = y/x for y/x small. 
                       #We cannot use Y because the vector is not of length 1. 
    for i in range (0,5)[::(-1)]: #i am adding the smallest angles first, since that makes addition marginally better.
            theta = theta + ArctanValues[10**(-i)]*ks[i]
    if CorrectSign == True:
        return theta
    else:
        return -theta
    
    
```

We test this out across values far from 0, values close to 0, and values very close to 0.

```python
for i in range(10):
    x = (5 - i)
    print(x, CordicArctan(x), math.atan(x))
for i in range(10):
    x = 1-i/5
    print(x, CordicArctan(x), math.atan(x))
for x in (-1000, -100, 100, 1000):
    print(x, CordicArctan(x), math.atan(x))
for x in (0.01, 0.001, 0.0001):
    print(x, CordicArctan(x), math.atan(x))
```

And we see that this function works very well, agreeing for 13 digits for most of the range, only loosing accuracy for arguments very close to zero, which we expect because our table only has stored arctan values down to $\arctan(0.0001)$.

## Question 4

To calculate arcsin and arccos we use the following facts:
$$\arcsin A = \arctan\frac{A}{\sqrt{1-A^2}}$$ $$\arccos A =  \pi/2 - \arcsin A$$
Thus I use CordicArctan in my implementation of CordicArcsin, and CordicArcsin in my implementation of CordicArccos

```python
def CordicArcsin(A):
    B = A/math.sqrt(1-A*A)
    return CordicArctan(B)
def CordicArccos(A):
    return math.pi/2 - CordicArcsin(A)
```
